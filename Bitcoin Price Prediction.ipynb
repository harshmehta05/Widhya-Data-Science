### Microtask 3

```python
correlation_mat = df.corr()
```

```python
print(correlation_mat['btc_market_price'])
```

btc_market_cap has perfect correlation with btc_market_price. Let us confirm this using jointplots

```python
col_to_correlate = df.columns
desired_output = 'btc_market_price'
```

```python
col_to_correlate
```

```python
import scipy.stats as stats

for i in range(col_to_correlate.size):
    if col_to_correlate[i] not in ['Date','btc_market_price'] :
        plot = sns.jointplot(col_to_correlate[i] ,desired_output, data=df, kind='reg')  
        plot.annotate(stats.pearsonr)
        plt.show()
```

Jointplots confirm that btc_market_cap has perfect correlation with btc_market_price (Refer jointplot number 2)

```python
from sklearn import metrics 
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
new_df = df[['btc_market_cap' , 'btc_n_transactions','btc_miners_revenue','btc_cost_per_transaction','btc_difficulty',
               'btc_hash_rate','btc_cost_per_transaction_percent','btc_market_price']]
```

```python
new_df.isnull().sum()
```

```python
new_df['btc_difficulty'].fillna(df['btc_difficulty'].median(), inplace = True)
```

```python
new_df.isnull().sum()
```

```python
new_df.shape
```

```python
X = new_df[['btc_market_cap' , 'btc_n_transactions','btc_miners_revenue','btc_cost_per_transaction','btc_difficulty',
               'btc_hash_rate','btc_cost_per_transaction_percent']]
```

```python
y = new_df['btc_market_price']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,random_state=2) 
```

```python
reg = LinearRegression() 
reg.fit(X_train, y_train)
```

```python
y_pred = reg.predict(X_test)
```

```python
metrics.mean_squared_error(y_test,y_pred)
```

```python
reg.score(X_test,y_test)
```
